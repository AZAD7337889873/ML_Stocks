```python
import pandas as pd
import numpy as np

np.random.seed(42)

rows = 10000

df = pd.DataFrame()

# Employee
df['EmployeeID'] = range(1, rows + 1)
df['Age'] = np.random.randint(18, 60, rows)

df['Gender'] = np.random.choice(
    ['Male','Female'],
    rows
)

df['MaritalStatus'] = np.random.choice(
    ['Single','Married','Divorced'],
    rows,
    p=[0.4,0.5,0.1]
)

df['EducationLevel'] = np.random.choice(
    ['High School','Bachelor','Master','PhD'],
    rows,
    p=[0.15,0.5,0.3,0.05]
)

df['EducationField'] = np.random.choice(
    ['IT','Business','Engineering',
     'Marketing','Finance','HR'],
    rows
)

# Job

df['Department'] = np.random.choice(
    ['IT','HR','Finance',
     'Sales','Operations'],
    rows
)

df['JobRole'] = np.random.choice(
    ['Analyst','Manager','Engineer',
     'Executive','Consultant'],
    rows
)

df['JobLevel'] = np.random.randint(1,6,rows)

df['TotalWorkingYears'] = np.random.randint(
    0,40,rows
)

df['YearsAtCompany'] = np.random.randint(
    0,20,rows
)

df['YearsInCurrentRole'] = np.random.randint(
    0,10,rows
)

df['YearsSinceLastPromotion'] = np.random.randint(
    0,10,rows
)

df['YearsWithCurrManager'] = np.random.randint(
    0,15,rows
)

# Compensation

df['MonthlyIncome'] = np.random.randint(
    20000,200000,rows
)

df['PercentSalaryHike'] = np.random.randint(
    5,30,rows
)

df['BonusPercent'] = np.random.randint(
    0,40,rows
)

df['StockOptionLevel'] = np.random.randint(
    0,4,rows
)

# Performance

df['PerformanceRating'] = np.random.randint(
    1,6,rows
)

df['TrainingTimesLastYear'] = np.random.randint(
    0,15,rows
)

df['CertificationCount'] = np.random.randint(
    0,20,rows
)

# Attendance

df['AbsenceDays'] = np.random.randint(
    0,30,rows
)

df['LateMarks'] = np.random.randint(
    0,20,rows
)

df['OvertimeHours'] = np.random.randint(
    0,100,rows
)

# Satisfaction

df['JobSatisfaction'] = np.random.randint(
    1,6,rows
)

df['EnvironmentSatisfaction'] = np.random.randint(
    1,6,rows
)

df['RelationshipSatisfaction'] = np.random.randint(
    1,6,rows
)

df['WorkLifeBalance'] = np.random.randint(
    1,6,rows
)

df['EmployeeEngagementScore'] = np.random.randint(
    1,101,rows
)

# Manager

df['ManagerRating'] = np.random.randint(
    1,6,rows
)

df['TeamSize'] = np.random.randint(
    2,25,rows
)

# Productivity

df['ProjectsHandled'] = np.random.randint(
    1,20,rows
)

df['AverageMonthlyHours'] = np.random.randint(
    120,320,rows
)

# Trends

df['PerformanceTrend3Months'] = np.random.normal(
    0,1,rows
)

df['PerformanceTrend6Months'] = np.random.normal(
    0,1,rows
)

df['SalaryGrowthLast3Years'] = np.random.randint(
    0,60,rows
)

# Create realistic attrition

risk_score = (
    (df['JobSatisfaction'] <= 2)*3 +
    (df['WorkLifeBalance'] <= 2)*3 +
    (df['AbsenceDays'] > 15)*2 +
    (df['OvertimeHours'] > 50)*2 +
    (df['YearsSinceLastPromotion'] > 5)*2 +
    (df['MonthlyIncome'] < 50000)*2 +
    (df['EmployeeEngagementScore'] < 40)*3
)

probability = 1/(1+np.exp(-risk_score))

df['Attrition'] = np.where(
    np.random.rand(rows) < probability*0.15,
    1,
    0
)

print(df.head())
```

```python
for col in ['MonthlyIncome',
            'JobSatisfaction',
            'ManagerRating']:

    df.loc[
        df.sample(frac=0.05).index,
        col
    ] = np.nan
```

```python
duplicates = df.sample(100)
df = pd.concat([df,duplicates])
```

```python
outlier_index = np.random.choice(
    df.index,
    100
)

df.loc[
    outlier_index,
    'MonthlyIncome'
] *= 5
```
